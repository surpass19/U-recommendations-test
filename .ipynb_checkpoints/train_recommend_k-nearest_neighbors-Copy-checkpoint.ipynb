# Recommendation building phase

## Collaborative Filtering

協調フィルタリングは、類似したユーザーの反応に基づいて、ユーザーが好む可能性のあるアイテムを除外できる手法です。 これは、大勢の人々を検索し、特定のユーザーに似た趣味を持つ少数のユーザーを見つけることで機能します。¶

## KNNを使用したコサイン類似性

コサイン類似度は、サイズに関係なく、ドキュメントがどの程度類似しているかを測定するために使用されるメトリックです。 数学的には、多次元空間に投影された2つのベクトル間の角度の余弦を測定します。 コサイン類似性は、2つの類似したドキュメントが（ドキュメントのサイズのために）ユークリッド距離だけ離れている場合でも、それらがより近くに向けられている可能性があるため、有利です。 角度が小さいほど、余弦の類似性が高くなります。

```python
from sklearn.neighbors import NearestNeighbors

model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(movie_pivot_sparse)
```

```python
query_index = np.random.choice(movie_pivot.shape[0])
distances, indices = model_knn.kneighbors(movie_pivot.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```

ここでは、ランダムに選択されたインデックス（movie_title）からKNNを介して6つの隣人の距離とインデックスを返します。これらは、推奨されるアニメのインデックスになります。

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}\n'.format(movie_pivot.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, movie_pivot.index[indices.flatten()[i]], distances.flatten()[i]))
```

# コンテンツベースのフィルタリング

コグニティブフィルタリングとも呼ばれるコンテンツベースのフィルタリングは、アイテムのコンテンツとユーザープロファイルの比較に基づいてアイテムを推奨します。 各アイテムのコンテンツは、記述子または用語のセット、通常はドキュメント内に出現する単語として表されます。コンテンツベースのレコメンダーは、ユーザーが明示的（評価）または暗黙的（リンクをクリック）に提供するデータを処理します。 そのデータに基づいて、ユーザープロファイルが生成され、ユーザーへの提案に使用されます。 ユーザーがより多くの入力を提供したり、推奨事項に対してアクションを実行したりすると、エンジンはますます正確になります。

## Term Frequency (TF) and Inverse Document Frequency (IDF)¶

TFは、単にドキュメント内の単語の頻度です。 IDFは、ドキュメントのコーパス全体におけるドキュメントの頻度の逆数です。 TF-IDFが使用される主な理由は、次の2つです。Googleで「分析の台頭」を検索するとします。 「the」は「analytics」よりも頻繁に発生することは確かですが、分析の相対的な重要性は検索クエリの観点よりも高くなります。 このような場合、TF-IDF均等化は、アイテム（ドキュメント）の重要性を判断する際の高頻度の単語の影響を打ち消します。
ここでは、ジャンルの内容に基づいてユーザーを推薦できるように、ジャンルで使用します。

```python
from sklearn.feature_extraction.text import TfidfVectorizer


tfv = TfidfVectorizer(min_df=3,  max_features=None, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')


# Filling NaNs with empty string
movie['genre'] = movie['genre'].fillna('')
genres_str = movie['genre'].str.split('|').astype(str)
tfv_matrix = tfv.fit_transform(genres_str)
```

Scikit-learnは、ベクトルコレクションの密な表現と疎な表現の両方で機能するペアワイズメトリック（機械学習用語ではカーネル）をすでに提供しています。 ここでは、推奨アニメに1を割り当て、非推奨アニメに0を割り当てる必要があります。 そのため、シグモイドカーネルを使用しています¶

```python
from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
indices = pd.Series(movie.index, index=movie['movie_title']).drop_duplicates()
```

## Recommendation function
ここでは、アニメのおすすめを取得するための関数を作成します。 列挙関数を使用して類似性スコアをリストに変換し、リストを並べ替えて、推奨する上位10のスコアを選択します。

```python
# データセットのタイトルをキーワードで検索
def searchanime(string):
    print(movie_pivot[movie_pivot.index.str.contains(string)].index[0:])
```

```python
searchanime('ゼロ')
```

```python
def give_rec(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]
    print(idx)

    # Get the pairwsie similarity scores 
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies 
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    movie_indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    return pd.DataFrame({'Movie name': movie['movie_title'].iloc[movie_indices].values,
                                 'Rating': movie['mean_review_point'].iloc[movie_indices].values})
```

```python
give_rec('名探偵コナン　ゼロの執行人')
```

# k近傍法（k-nearest neighbors）でレコメンド機能を構築

k近傍法ですが最も単純なアルゴリズムと呼ばれており、クラスタリングのアルゴリズムの一種で、一般的に分類問題などで利用されます。今回のチュートリアルの例で考えると、とあるアニメのタイトルを入力した際に、そのアニメのneighbors（ご近所さん）を探して、距離が近い（類似性が高い）アニメのタイトルをオススメとして返してくれます。

```python
# Scikit-learnのライブラリを利用します
# n_neiborsやalgorithm、metricなど重要なアーギュメントを設定しています
knn = NearestNeighbors(n_neighbors=9,algorithm= 'brute', metric= 'cosine')
 
# 前処理したデータセットでモデルを訓練
model_knn = knn.fit(movie_pivot_sparse)
```

```python
# データセットのタイトルをキーワードで検索
def searchanime(string):
    print(movie_pivot[movie_pivot.index.str.contains(string)].index[0:])
```

```python
searchanime('イエス')
```

```python
# 「はじめの一歩」に対してのオススメのアニメ10個
Movie = '名探偵コナン　ゼロの執行人'
 
distance, indice = model_knn.kneighbors(movie_pivot.iloc[movie_pivot.index== Movie].values.reshape(1,-1),n_neighbors=20)

for i in range(0, len(distance.flatten())):
    if  i == 0:
        print('Recommendations if you like the anime {0}:\n'.format(movie_pivot[movie_pivot.index== Movie].index[0]))
    else:
        print('{0}: {1} with distance: {2}'.format(i,movie_pivot.index[indice.flatten()[i]],distance.flatten()[i]))
```

```python
# 「はじめの一歩」に対してのオススメのアニメ10個
Movie = '最後の恋、初めての恋'
 
distance, indice = model_knn.kneighbors(movie_pivot.iloc[movie_pivot.index== Movie].values.reshape(1,-1),n_neighbors=11)

for i in range(0, len(distance.flatten())):
    if  i == 0:
        print('Recommendations if you like the anime {0}:\n'.format(movie_pivot[movie_pivot.index== Movie].index[0]))
    else:
        print('{0}: {1} with distance: {2}'.format(i,movie_pivot.index[indice.flatten()[i]],distance.flatten()[i]))
```
