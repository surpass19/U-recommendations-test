```python
#みんなのシネマレビュー
url = "https://www.jtnews.jp/cgi-bin/revlist.cgi?PAGE_NO=137"
res = requests.get(url)
res_bs = bs(res.content, "html.parser")
```

## 関数化

```python
def make_use_df(res_bs):
    output = []

    use_list = res_bs.find(class_="hover-table").find_all("a")

    for user in use_list:
        user_url = user.get("href")
        user_name = user.text
        user_id = user_pattern.findall(user_url)[0]
        
        output.append({
            "user_id":user_id,
            "user_name":user_name,
        })

    df = pd.DataFrame(output)
    
    return df
```

## スクレイピング

```python
all_df = pd.DataFrame()

for num in tqdm(range(1, 138)):
    #urlの生成 #みんなのシネマレビュー
    url = f"https://www.jtnews.jp/cgi-bin/revlist.cgi?PAGE_NO={num}"
    #スクレイピング
    res = requests.get(url)
    res_bs = bs(res.content, "html.parser")
    
    df = make_use_df(res_bs)
    all_df = all_df.append(df, ignore_index=True)
    time.sleep(2)

all_df = all_df[all_df.user_name != '--------']    
all_df
```

## 関数化

```python
def get_html(url):
    headers = {"User-Agent": useragent}
    res = requests.get(url, headers=headers)
    return res
```

```python
def make_movie_dic_df(res_bs, dic_movie):
    output = []

    table = res_bs.findAll("table", {"class":"review-area"})[0]
    rows = table.findAll("tr")

    for row in rows:
        comment = row.find("div").text.replace("\xa0","")
        movie_tile = row.findAll("a")[1].text.replace(" ","")
        row_url = row.find("a").get("href")
        
        user_id = int(user_pattern.findall(row_url)[0])
        movie_id = int(movie_pattern.findall(row_url)[0])
        row_text = row.text
        point = int(point_pattern.findall(row_text)[-1].replace("点（", ""))

        dic_movie.setdefault(str(movie_id), movie_tile)
        output.append({
            "user_id":user_id,
            "movie_id":movie_id,
            "movie_tile":movie_tile,
            "movie_comment":comment,
            "point":point,
        })

    return pd.DataFrame(output), dic_movie
```

```python
def get_review_page_count(res_bs):
    #reviewのページ数を取得する関数
    pagging_table = res_bs.findAll("table", {"class":"pagging-table"})[0]
    #レビュー(評価順, 投稿日付順, 変更日付順)のページ数の列数
    pagging_rows = pagging_table.findAll("tr")
    #レビュー(評価順)の列数に絞り込み
    review_rows = int(len(pagging_rows)/3)
    
    #レビューのページ数
    if review_rows == 0:
        review_page_count = 0
    else:
        review_page_count = int(pagging_rows[review_rows-1].findAll("a")[-1].text)
    
    return review_page_count
```

## 関数化

```python
regex = re.compile(r"\d{1,4}")
hun = re.compile('\d+')

def make_movie_df(res_bs):
    output = []

    review_point_class = res_bs.find(class_="review-point")
    if review_point_class is not None:
        review_point = float(review_point_class.text.split("：")[1].split(" / ")[0])
    else:
        review_point = np.nan

    review_count_class = res_bs.find(class_="review-count")
    if review_count_class is not None:
        review_num = int(regex.findall(review_count_class.text)[0])
    else:
        review_num = np.nan
    
    a_tags = res_bs.findAll("a")
    a_tag_list=[]
    for a_tag in a_tags:
        if "_KIND=4" in str(a_tag):
            a_tag_list.append(a_tag.text)
    genre = "|".join(a_tag_list)
    
    span = res_bs.find("span",{"style":"color:#60f"})
    if span is not None:
        screening_time = int(hun.findall(span.text)[0])
    else:
        screening_time = np.nan

    output.append({
                "mean_review_point":review_point,
                "number_of_revier":review_num,
                "screening_time":screening_time,
                "genre":genre,
            })

    df = pd.DataFrame(output)

    return df
```
